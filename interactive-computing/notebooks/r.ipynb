## 2. A small tibble

A `tibble` is the tidyverse's friendly version of a data frame — same idea as a pandas DataFrame, with nicer printing.

```r
suppressPackageStartupMessages({
  library(tibble)
  library(dplyr)
})

df <- tibble(
  sample_id      = c("S01", "S02", "S03", "S04", "S05"),
  condition      = c("control", "control", "treated", "treated", "treated"),
  reads_millions = c(42.1, 39.8, 44.6, 41.0, 38.2)
)
df
```

```r
# Group-and-summarise — read counts averaged by condition.
df %>%
  group_by(condition) %>%
  summarise(mean_reads = mean(reads_millions))
```

## 3. A first plot

`ggplot2` is R's grammar-of-graphics library. The cell below produces the same bar chart we made on the Python side.

```r
suppressPackageStartupMessages(library(ggplot2))

summary_df <- df %>%
  group_by(condition) %>%
  summarise(mean_reads = mean(reads_millions))

ggplot(summary_df, aes(x = condition, y = mean_reads, fill = condition)) +
  geom_col(width = 0.6) +
  scale_fill_manual(values = c(control = "steelblue", treated = "indianred")) +
  labs(y = "reads (millions)", title = "Mean reads per condition") +
  theme_minimal(base_size = 12) +
  theme(legend.position = "none")
```

## 4. Inspect your environment

Save the output of the next cell. When a notebook stops working months from now, this is the first thing you'll want — the actual versions you were running.

```r
cat("R version:    ", R.version$version.string, "\n")
cat("Platform:     ", R.version$platform, "\n")
cat("Working dir:  ", getwd(), "\n")
conda_env <- Sys.getenv("CONDA_DEFAULT_ENV")
cat("Conda env:    ", ifelse(nzchar(conda_env), conda_env, "(not in conda)"), "\n\n")

for (pkg in c("tibble", "dplyr", "ggplot2", "Seurat")) {
  ver <- tryCatch(
    as.character(packageVersion(pkg)),
    error = function(e) "(not installed)"
  )
  cat(sprintf("%-12s %s\n", pkg, ver))
}
```
