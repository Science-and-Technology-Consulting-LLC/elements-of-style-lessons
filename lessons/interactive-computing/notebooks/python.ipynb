## 2. A small DataFrame

`pandas` is the Python library you'll meet first for tabular data. A DataFrame is a table — rows and columns, with labels.

```python
import pandas as pd

df = pd.DataFrame({
    "sample_id": ["S01", "S02", "S03", "S04", "S05"],
    "condition": ["control", "control", "treated", "treated", "treated"],
    "reads_millions": [42.1, 39.8, 44.6, 41.0, 38.2],
})
df
```

```python
# A typical "summary by group" — read counts averaged by condition.
df.groupby("condition")["reads_millions"].mean()
```

## 3. A first plot

`matplotlib` is the workhorse plotting library. The cell below produces a small bar chart of the same summary.

```python
import matplotlib.pyplot as plt

summary = df.groupby("condition")["reads_millions"].mean()

fig, ax = plt.subplots(figsize=(4, 3))
summary.plot(kind="bar", ax=ax, color=["steelblue", "indianred"])
ax.set_ylabel("reads (millions)")
ax.set_title("Mean reads per condition")
plt.tight_layout()
plt.show()
```

## 4. Inspect your environment

When something stops working in a notebook six months from now, the *first* thing you'll want is the answer to **"what was I running?"** This cell prints it.

Save its output. Paste it into a bug report. Future you will thank present you.

```python
import sys, platform, os

print("Python:        ", sys.version.split()[0])
print("Executable:    ", sys.executable)
print("Platform:      ", platform.platform())
print("Conda env:     ", os.environ.get("CONDA_DEFAULT_ENV", "(not in conda)"))
print("Working dir:   ", os.getcwd())
print()

for pkg in ["pandas", "numpy", "matplotlib", "scanpy"]:
    try:
        mod = __import__(pkg)
        print(f"{pkg:14s} {getattr(mod, '__version__', '?')}")
    except ImportError:
        print(f"{pkg:14s} (not installed)")
```
